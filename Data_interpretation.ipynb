```python
def mean_transmission(time): 
    data_before = pd.read_csv (r'E:/LIAD_DATA_PON/16_2_2565/data__16_2_2565_data_liad_'+str(time)+'sec_before_absorb.csv') 
    data_after = pd.read_csv (r'E:/LIAD_DATA_PON/16_2_2565/data__16_2_2565_data_liad_'+str(time)+'sec_after_absorb.csv') 
    data_before["after"]=data_after["0"]   
    data = data_before.rename({'Unnamed: 0': 'points', '0': 'before'}, axis=1) 
    xincr=1e-06
    v_max_before=data["before"].max()  
    v_min_before=data["before"].min()   
    data['before']=((data['before']-v_min_before)*100)/(v_max_before-v_min_before) 
    v_max_after=data["after"].max()  
    v_min_after=data["after"].min()   
    data['after']=((data['after']-v_min_after)*100)/(v_max_after-v_min_after) 
    data['points']=data['points']*xincr  
    data['transmission']= data['after']-data['before']
    return data['transmission'].mean()
    
```

```python
data_time = np.arange(1,30.5,0.5)  
transmission_dependence_on_time = []
for period in data_time: 
    transmission_dependence_on_time.append(mean_transmission(period)) 
table_transmission=pd.DataFrame(data_time,columns =['interval time'])   
table_transmission['transmission']=pd.DataFrame(transmission_dependence_on_time) 
```

```python
import numpy as np
import pylab as plt
import itertools

k = 20
X = np.linspace(0, 20, 500)
Y = np.zeros((k,X.size))

# Add some fake data
MU = np.random.random(k)
for n in xrange(k):
    Y[n] += np.exp(-(X-MU[n]*n)**2 / (1+n/3))
Y *= 50

# Add some outliers for show
Y += 2*np.random.random(Y.shape)

displace = Y.max()/4

# Add a cutoff
Y[Y<1.0] = None

face_colors = itertools.cycle(["#D3D820", "#C9CC54", 
                               "#D7DA66", "#FDFE42"])

fig = plt.figure()
ax = fig.add_subplot(111, axisbg='black')
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

for n,y in enumerate(Y):
    # Vertically displace each plot
    y0 = np.ones(y.shape) * n * displace
    y1 = y + n*displace

    plt.fill_between(X, y0,y1,lw=1, 
                     facecolor=face_colors.next(),
                     zorder=len(Y)-n)  
plt.show()
```
